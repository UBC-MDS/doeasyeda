```python
import altair as alt
import pandas as pd

from doeasyeda.create_scatter_plot import create_scatter_plot
from doeasyeda.create_line_plot import create_line_plot
from doeasyeda.create_hist_plot import create_hist_plot
from doeasyeda.create_area_plot import create_area_plot
```

```python
df = pd.read_csv('gapminder.csv')
df.head()
```

```python
create_scatter_plot(df, 'continent', 'lifeExp', color='continent', title='Life Exp by Continent', x_title= 'Continent', y_title='Life Exp')
```

```python
df_grouped1 = df.groupby(['continent'])['lifeExp'].sum().reset_index()
create_hist_plot(df_grouped1, 'continent', 'lifeExp', color='continent', title='Average Life Exp by Continent', x_title= 'Continent', y_title='Average Life Exp')
```

```python
df_grouped2 = df.groupby(['continent', 'year'])['population'].sum().reset_index()
create_area_plot(df_grouped2, 'year', 'population', color='continent', title='Total Population by Continent', x_title= 'Continent', y_title='Total Population')
```

```python
df['gdp'] = df['gdpPercap'] * df['population']
df_grouped3 = df.groupby(['continent', 'year'])[['population', 'gdp']].sum().reset_index()
df_grouped3['gdpPercap'] = df_grouped3['gdp']/df_grouped3['population']

create_line_plot(df_grouped3, 'year', 'gdpPercap', color='continent', title=' GDP per capita by Continent', x_title= 'Continent', y_title='GDP per capita')
```
